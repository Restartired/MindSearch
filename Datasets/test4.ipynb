# Import Required Libraries
Import the necessary libraries, such as `json` and `os`.

```python
# Import Required Libraries
import json
import os
```

# Load JSON Data
Load the JSON data from the file `/root/repo/MindSearch/Datasets/mindsearch_results_5.json` using the `json` module.

```python
# Load JSON Data
file_path = "/root/repo/MindSearch/Datasets/mindsearch_results_5.json"

# Check if the file exists
if not os.path.exists(file_path):
    raise FileNotFoundError(f"The file {file_path} does not exist.")

# Load the JSON data
with open(file_path, "r", encoding="utf-8") as f:
    data = json.load(f)
```

# Extract Questions with Empty Responses
Iterate through the data and extract questions where the `responses` field is an empty list.

```python
# Extract Questions with Empty Responses
empty_responses = [item["query"] for item in data if item["responses"] == []]

# Count the number of questions with empty responses
empty_responses_count = len(empty_responses)
total_questions = len(data)

# Calculate the ratio of empty responses
empty_responses_ratio = empty_responses_count / total_questions if total_questions > 0 else 0
```

# Display and Save Results
Print the extracted questions and save them to a new JSON file, such as `empty_responses.json`.

```python
# Display Results
print(f"Total Questions: {total_questions}")
print(f"Questions with Empty Responses: {empty_responses_count}")
print(f"Ratio of Empty Responses: {empty_responses_ratio:.2%}")
print("Questions with Empty Responses:")
for query in empty_responses:
    print(f"- {query}")

# Save Results to a JSON File
output_file = "/root/repo/MindSearch/Datasets/empty_responses.json"
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(empty_responses, f, ensure_ascii=False, indent=4)

print(f"Extracted questions saved to {output_file}")
```
